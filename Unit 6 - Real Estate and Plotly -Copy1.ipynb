```python
# neighborhoods=pd.read_excel('/Users/chris/Downloads/2019.1.15__SF_ASR_Secured_Roll_Data_2017-2018.xlsx',
#                            sheetname='Neighborhood Code')
neighborhoods=pd.read_excel('/Users/nigel/Downloads/2019.1.15__SF_ASR_Secured_Roll_Data_2017-2018.xlsx',
                           sheetname='Neighborhood Code')
neighborhoods['Neighborhood Codes']=(('0'+neighborhoods['CODE']).str[-3:])
neighborhoods.head()
```

```python
#Read in all tax-year assessor data
import os as os
# Concats all years in tax data directory to calculate neighborhood appreciation
# sf_tax_path = "/Users/chris/Dropbox/Trilogy/Real Estate/tax_data/"
sf_tax_path = "/Users/nigel/SFPropertyTax/data/"
df = pd.concat([pd.read_csv(sf_tax_path + df, low_memory=False) for df in os.listdir(sf_tax_path)])
df["RP1LNDVAL"] = df["RP1LNDVAL"].astype(float)

# Only select taxed values > $0
df = df[df["RP1LNDVAL"] > 0]

```

```python
df.head()
```

```python
df['ROLLYEAR'].drop_duplicates()
```

```python
# Convert decimal year to string
df['ROLLYEAR'] = (2000 + df['ROLLYEAR']).astype(str).str[:-2].astype(float)
df['ROLLYEAR'].drop_duplicates()

        
```

```python
df.head()
```

```python

df['RP1PPTVAL']=df['RP1PPTVAL'].replace(0, np.nan)
df.dropna(subset=['RP1PPTVAL'], inplace=True)
# df['SaleDate']
df['RECURRSALD']=df['RECURRSALD'].astype(str)
df['RECURRSALD'].replace('.', '', inplace=True)
df.loc[df['RECURRSALD']=='0', 'RECURRSALD' ] = np.nan
# df['Sale Year']=''
df['Sale Year']=0
df['Sale Digit']=pd.to_numeric(df['RECURRSALD'].str[0:2], errors='coerce')
# df.head()
df.reset_index(inplace=True)
# Fixing 'Sale Year' column
df.loc[df['Sale Digit']<=17, 'Sale Year']=2000+df['Sale Digit']
df.loc[df['Sale Digit']>17, 'Sale Year']=1900+df['Sale Digit']
# df['Sale Digit']=df['Sale Digit'].astype(int)
# df['Sale Year']=df['Sale Year'].astype(int)
# df
df.loc[df['Sale Year']>2017, 'Sale Year']=np.nan
df['Assessed Value']=df['RP1FXTVAL']+df['RP1IMPVAL']+df['RP1LNDVAL']
df['Years Since Sale']=2017-df['Sale Year']

# Proposition 13, 2% cap
df['Sale Price']=(df['Assessed Value']/((1.02)**df['Years Since Sale']))
df['Sale Year']=df['Sale Year']
df['Sale Price Per Square Foot']=df['Sale Price']/df['SQFT']
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df
```

```python
df['District']=df['RP1NBRCDE'].astype(str).str[0:2]
df['ROLLYEAR'].drop_duplicates()
```

```python
#Neighbborhood Info
#Merge in the names for each neighborhood
df=df.merge(neighborhoods, left_on=['RP1NBRCDE'], right_on=['Neighborhood Codes'])
df.head()
```

```python
# df['PROPLOC'].str.contain( 'MAILADRS3') 
df=df[df['MAILADRS4']=='SAN FRANCISCO CA']
df.head()
```

```python
business=business[business['City']=='San Francisco']
```

```python
# business.groupby(by=['Source Zipcode'])
for x in ['Business Start Date', 'Business End Date', 'Location Start Date', 'Location End Date']:
    business[x]=pd.to_datetime(business[x], infer_datetime_format=True)
business.head()
```

```python
business['Start Year']=business['Business Start Date'].dt.year
business.head()
```

```python

business_activity=business[business['Start Year']>=2009].groupby(by=['Start Year', 'Source Zipcode'])[['Source Zipcode']].count()
business_activity.columns=['# of businesses']
business_activity['lag bus']=business_activity.groupby(by=['Source Zipcode'])['# of businesses'].shift()
business_activity['Business Growth']=business_activity['# of businesses']/business_activity['lag bus']-1
business_activity['Business Growth'].unstack(level=1).plot(kind='bar', stacked=True, legend=False)
plt.show()
business_activity['# of businesses'].unstack(level=1).plot(kind='bar', stacked=True, legend=False)
business_activity=business_activity[['# of businesses', 'Business Growth']]
business_activity.reset_index(inplace=True)
business_activity.rename(columns={'Source Zipcode':'Zip Code', 'Start Year':'Business Year'}, inplace=True)
business_activity['Zip Code'] = business_activity['Zip Code'].astype(int)
business_activity
```

```python
import zipcodes
business_activity['Zip Code'] = business_activity['Zip Code'].astype(str)
zips = zipcodes.matching(business_activity.loc[0,'Zip Code'])
zips[0]['lat']
```

```python
import zipcodes

business_activity['Zip Code'] = business_activity['Zip Code'].astype(str).str[0:5]

business_activity['lat']=np.nan
business_activity['lon']=np.nan
for idx, row in business_activity.iterrows():
    try:
        business_activity.loc[idx, 'lat']=zipcodes.matching(row['Zip Code'])[0]['lat']
    except:
        pass
    try:
        business_activity.loc[idx, 'lon']=zipcodes.matching(row['Zip Code'])[0]['long']
    except:
        pass

business_activity.head()
```

```python
from sklearn.preprocessing import scale
from sklearn import preprocessing
        
min_max_scaler = preprocessing.MinMaxScaler()
business_activity.dropna(subset=['Business Growth'], inplace=True)
business_activity['New Business Growth - Scaled'] = scale(business_activity['Business Growth'])
business_activity.head()
```
